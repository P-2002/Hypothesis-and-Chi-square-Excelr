**CHI-SQUARE TEST**

**State the Hypothesis**


*   Null Hypothesis (H0) : There is no association between the type of smart home devices (Smart Thermostat Vs Smart Light) and the customer's satisfaction level.
*  Alternative Hypothesis (Ha) : There is an association between the type of smart home devices and the customer's satisfaction level.



**Compute the chi-square statistic**

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency,chi2

#data
data=[[50, 70], [80, 100], [60, 90], [30, 50], [20, 50]]
df=pd.DataFrame(data,
             index=['Very Satisfied','Satisfied','Neutral','Unsatisfied','Very Unsatisfied'], # row names
             columns=['Smart Thermostat','Smart Light']) # column names
df
```

```python
chisquare, PVALUE, dof, expected = chi2_contingency(data)
print("Chi-Square Statistic:", chisquare.round(4))
print("P-Value:", PVALUE.round(4))
print("Degree of Freedom:", dof)
print("Expected Frequencies:", expected)
```

**Determine the critical-value**

```python
alpha=0.05
critical_value = chi2.ppf(1 - alpha, dof)
print("Critical Value:", critical_value.round(4))
chi2_statistic = 10.2
```

```python
if chi2_statistic > critical_value:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis. ")
```
